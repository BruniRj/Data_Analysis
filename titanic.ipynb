```python
# utilisation de seaborn pour une visualisation 
# Charge le fichier titanic.csv depuis Kaggle (ou utilise seaborn.load_dataset("titanic")).
# Questions :

# Combien de passagers ?

# Âge moyen des survivants ?

# Nombre d'hommes et de femmes ?
import seaborn as sns 
titanic = sns.load_dataset("titanic")
print(titanic.head())

# nombre de passager : 
passager = len(titanic)
print(f"voici le nombre de passager : {titanic.shape[0]}") #nb de ligne seulement parceque .shape retourne un tuple avec ligne et colonne donc si on fait .shape seulement on a 891 passagers qui est la ligne et 15 colonne qui sont les information du passager 

# age moyen des survivant 
ageMoyen = titanic[titanic["survived"]==1]["age"].mean()
print(f"voici l'age moyen : {ageMoyen:.2f}")

# Nombre dhomes et femmes 
# c’est une Series Pandas. Par défaut, Pandas affiche non seulement les valeurs mais aussi des métadonnées (le nom et le type).
# il nous suffit juste d'ajouter .to_dict() par exemple pour ne pas retourner le nom et le type  to_dict
print(titanic["sex"].value_counts().reset_index())
```
